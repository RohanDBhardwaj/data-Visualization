```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10,50,30,40]
min_value = min(y)
max_value = max(y)
plt.title("example for checking working")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.axhline(y=min_value, color='b', linestyle='-', label='Min Value')
plt.axhline(y=max_value, color='g', linestyle='-', label='Max Value')
plt.plot(x,y,color="red",linestyle='--',linewidth = '2',label='no. of pt' )
plt.legend(loc='best')
plt.show()
```

```python
data2=pd.read_csv('countries.csv')
data2
```

```python
us = data2[data2.country=="United States"]
china = data2[data2.country=="China"]
plt.plot(us.year,us.population / 10**6,label='us')
plt.plot(china.year,china.population / 10**6,label = 'china')
plt.xlabel('Years')
plt.ylabel('population M')
plt.legend(loc= 'best')
```

```python
us.population / us.population.iloc[0] * 100
```

```python
# us = data2[data2.country=="United States"]
# china = data2[data2.country=="China"]
plt.plot(us.year,us.population / us.population.iloc[0] * 100,label='us')
plt.plot(china.year,china.population / china.population.iloc[0] * 100,label = 'china')
plt.xlabel('Years')
plt.ylabel('population growth M(first year = 100m)')
plt.legend(loc= 'best')
```

```python
india = data2[data2.country=="India"]
india
```
